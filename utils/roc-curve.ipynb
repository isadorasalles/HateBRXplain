```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from matplotlib import rc 

rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
rc('text.latex', preamble=r'\usepackage{amsmath}\usepackage{amssymb}')
plt.rcParams['text.latex.preamble'] = r'\usepackage{color}'
plt.rcParams["figure.figsize"] = (6,5)
plt.rcParams['savefig.dpi'] = 300  # Ensures high-quality output
plt.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_csv('../data/hatebr_and_rationales.csv', \
                index_col=0, \
                converters={"rationales_offensive_1_normalized": \
                lambda x: x.strip('[]').replace("'", "").split(", "),
                "rationales_offensive_2_normalized": \
                lambda x: x.strip('[]').replace("'", "").split(", ")})


## split to train and val
TRAIN_SIZE = 0.8
TEST_SIZE = 0.1
VAL_SIZE = 0.1

x_train, x_test_val, y_train, y_test_val = train_test_split(df['normalized_text'], df['label final'], test_size=TEST_SIZE + VAL_SIZE, random_state=0)
x_test, x_val, y_test, y_val = train_test_split(x_test_val, y_test_val, test_size=VAL_SIZE/(TEST_SIZE + VAL_SIZE), random_state=0)

y_true = y_test
```

```python

plt.figure()

y_pred = np.loadtxt('../predictions/mbert_pred.out')

fpr, tpr, thresholds = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='y', lw=2, label='mBERT (AUC = %0.4f)' % roc_auc)


y_pred1 = np.loadtxt('../predictions/bertimbau_pred.out')

fpr, tpr, thresholds = roc_curve(y_true, y_pred1)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='c', lw=2, label='BERTimbau (AUC = %0.4f)' % roc_auc)


y_pred2 = np.loadtxt('../predictions/distil_pred.out')

fpr, tpr, thresholds = roc_curve(y_true, y_pred2)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='tab:purple', lw=2, label='DistilBERTimbau (AUC = %0.4f)' % roc_auc)


y_pred3 = np.loadtxt('../predictions/ptt5_pred.out')

fpr, tpr, thresholds = roc_curve(y_true, y_pred3)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='darkred', lw=2, label='PTT5 (AUC = %0.4f)' % roc_auc)


plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=20)
plt.ylabel('True Positive Rate',  fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
# plt.title('Receiver Operating Characteristic (ROC)', fontsize=18)

legend = plt.legend(loc='lower right', bbox_to_anchor=(1.5, 0), fontsize=18)
plt.setp(legend.get_texts(), color='black') 
# plt.legend(loc="lower right", fontsize=16)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(True)
plt.gca().spines['bottom'].set_visible(True)

plt.gca().spines['left'].set_linewidth(0.6)
plt.gca().spines['bottom'].set_linewidth(0.6)

plt.gca().spines['left'].set_color('grey')
plt.gca().spines['bottom'].set_color('grey')

plt.savefig('roc_curves.jpg', bbox_inches = 'tight')
plt.show()
```
